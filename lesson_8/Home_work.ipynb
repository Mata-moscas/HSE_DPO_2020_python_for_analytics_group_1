Помни - данные это конечно хорошо, но без твоих аналитических навыков мы не как не справимся. Потому нас будет интересовать вопросы примерное в следующей сфере:
- как устренны скидки? в каком диапазоне бывают? какая доля товаров со скидками?
- какие бренды самые популярные? почему похоже что они популярные?
- все ли всегда есть в наличии? какие товары отсутствуют? какая их доля и в каких категориях их больше всего? может с них и начнем?
- отдельно интересуют дорогие товары. большой оборот - это конечно хорошо, но... давай честно, нас интересует маржа. давай посмотрим на самые дорогие товары, если есть оценки/отзовы - насколько они отличаются от обычных? как много вообще премиум товаров относительно обычных?
